```python
import pandas as pd
```

```python
data = pd.read_json('./weaponStats.json', orient='record')
data.head()
```

```python
list(data)
```

```python
list(data['Type'].unique())
```

```python
filter = data['Type'] == 'Glaive'
glaives = data[filter].copy()
glaives.dropna(axis=1, inplace=True, how='all')
display(glaives.isna().all())
display(glaives.head())
glaives.shape
```

```python
filter = data['Type'] == 'Auto Rifle'
auto_rifles = data[filter].copy()
auto_rifles.dropna(axis=1, inplace=True, how='all')
display(auto_rifles.isna().all())
display(auto_rifles.head())
auto_rifles.shape
```

```python
filter = data['Type'] == 'Combat Bow'
bows = data[filter].copy()
bows.dropna(axis=1, inplace=True, how='all')
display(bows.isna().all())
display(bows.head())
bows.shape
```

```python
filter = data['Type'] == 'Fusion Rifle'
fusion_rifles = data[filter].copy()
fusion_rifles.dropna(axis=1, inplace=True, how='all')
display(fusion_rifles.isna().any())
display(fusion_rifles.head())
fusion_rifles.shape
```

```python
filter = data['Type'] == 'Grenade Launcher'
grenade_launchers = data[filter].copy()
grenade_launchers.dropna(axis=1, inplace=True, how='all')
display(grenade_launchers.isna().any())
display(grenade_launchers.head())
grenade_launchers.shape
```

```python
filter = data['Type'] == 'Hand Cannon'
hand_cannons = data[filter].copy()
hand_cannons.dropna(axis=1, inplace=True, how='all')
display(hand_cannons.isna().any())
display(hand_cannons.head())
hand_cannons.shape
```

```python
filter = data['Type'] == 'Linear Fusion Rifle'
linear_frs = data[filter].copy()
linear_frs.dropna(axis=1, inplace=True, how='all')
display(linear_frs.isna().any())
display(linear_frs.head())
linear_frs.shape
```

```python
filter = data['Type'] == 'Pulse Rifle'
pulse_rifles = data[filter].copy()
pulse_rifles.dropna(axis=1, inplace=True, how='all')
display(pulse_rifles.isna().any())
display(pulse_rifles.head())
pulse_rifles.shape
```

```python
filter = data['Type'] == 'Scout Rifle'
scout_rifles = data[filter].copy()
scout_rifles.dropna(axis=1, inplace=True, how='all')
display(scout_rifles.isna().any())
display(scout_rifles.head())
scout_rifles.shape
```

```python
filter = data['Type'] == 'Shotgun'
shotguns = data[filter].copy()
shotguns.dropna(axis=1, inplace=True, how='all')
display(shotguns.isna().any())
display(shotguns.head())
shotguns.shape
```

```python
filter = data['Type'] == 'Sidearm'
sidearms = data[filter].copy()
sidearms.dropna(axis=1, inplace=True, how='all')
display(sidearms.isna().any())
display(sidearms.head())
sidearms.shape
```

```python
filter = data['Type'] == 'Sniper Rifle'
sniper_rifles = data[filter].copy()
sniper_rifles.dropna(axis=1, inplace=True, how='all')
display(sniper_rifles.isna().any())
display(sniper_rifles.head())
sniper_rifles.shape
```

```python
filter = data['Type'] == 'Submachine Gun'
submachine_guns = data[filter].copy()
submachine_guns.dropna(axis=1, inplace=True, how='all')
display(submachine_guns.isna().any())
display(submachine_guns.head())
submachine_guns.shape
```

```python
filter = data['Type'] == 'Trace Rifle'
trace_rifles = data[filter].copy()
trace_rifles.dropna(axis=1, how='all', inplace=True)
display(trace_rifles.isna().any())
display(trace_rifles.head())
trace_rifles.shape
```

```python
filter = data['Type'] == 'Machine Gun'
machine_guns = data[filter].copy()
machine_guns.dropna(axis=1, how='all', inplace=True)
display(machine_guns.isna().any())
display(machine_guns.head())
machine_guns.shape
```

```python
filter = data['Type'] == 'Rocket Launcher'
rocket_launchers = data[filter].copy()
rocket_launchers.dropna(axis=1, how='all', inplace=True)
display(rocket_launchers.isna().any())
display(rocket_launchers.head())
rocket_launchers.shape
```

```python
filter = data['Type'] == 'Sword'
swords = data[filter].copy()
swords.dropna(axis=1, how='all', inplace=True)
display(swords.isna().any())
display(swords.head())
swords.shape
```
